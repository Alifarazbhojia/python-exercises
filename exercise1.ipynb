```python
# Q1
def string_ops(s):
    return s[::-1], s[::2]

# Q2
nums = [1,2,3,4,5,6]
even_squares = [x**2 for x in nums if x % 2 == 0]

# Q3
def char_freq(s):
    freq = {}
    for c in s:
        freq[c] = freq.get(c,0)+1
    return freq

# Q4
def odd_indices(lst):
    return [i for i, x in enumerate(lst) if x % 2 != 0]

# Q5
def rotate(lst, k):
    return lst[-k:] + lst[:-k]

# Q6
def list_to_dict(a,b):
    return dict(zip(a,b))

# Q7
def manual_iter(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            break

# Q8
def flatten(lst):
    return [item for sublist in lst for item in sublist]

# Q9
def is_palindrome(s):
    return s == s[::-1]

# Q10
def second_largest(lst):
    first = second = float('-inf')
    for n in lst:
        if n > first:
            second = first
            first = n
        elif n > second and n != first:
            second = n
    return second
```
